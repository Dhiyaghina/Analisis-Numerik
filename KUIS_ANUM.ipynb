1. Orde 3

```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan diferensial dy/dx = 2
def f(x, y):
    return 2

def runge_kutta_3(x0, y0, h, x_target):
    x = x0
    y = y0
    while x < x_target:
        k1 = h * f(x, y)
        k2 = h * f(x + h/2, y + k1/2)
        k3 = h * f(x + h, y - k1 + 2*k2)
        y = y + (k1 + 4*k2 + k3) / 6
        x = x + h
    return y

# Awal

x0 = 0.0    # Nilai awal x
y0 = 0.0    # Nilai awal y
x_target = 3.0  # Target x
h = 0.3     # Ukuran langkah

# Menghitung ùë¶(3.0) dengan Runge-Kutta orde 3
y = runge_kutta_3(x0, y0, h, x_target)

# Mencetak hasil perhitungan ùë¶(3.0)
print(f"Hasil Runge-Kutta orde 3: ùë¶(3.0) = {y}")

# Membuat Grafik Numerik
x_values = np.arange(x0, x_target + h, h)
y_values = []

# Menghitung nilai y pada setiap langkah x dan menyimpannya dalam y_values
for x in x_values:
    y_values.append(runge_kutta_3(x0, y0, h, x))

# Menampilkan Hasil Grafik Numerik
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label="Runge-Kutta Orde 3", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hasil Grafik Orde 3")
plt.legend()
plt.grid(True)
plt.show()
```

1. Orde 4

```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan diferensial dy/dx = 2
def f(x, y):
    return 2

def runge_kutta_4(x0, y0, h, x_target):
    x = x0
    y = y0
    while x < x_target:
        k1 = h * f(x, y)
        k2 = h * f(x + h/2, y + k1/2)
        k3 = h * f(x + h/2, y + k2/2)
        k4 = h * f(x + h, y + k3)
        y = y + (k1 + 2*k2 + 2*k3 + k4) / 6
        x = x + h
    return y

# Awal

x0 = 0.0    # Nilai awal x
y0 = 0.0    # Nilai awal y
x_target = 3.0  # Target x
h = 0.3     # Ukuran langkah

# Menghitung ùë¶(3.0) dengan Runge-Kutta orde 4
y = runge_kutta_4(x0, y0, h, x_target)

# Mencetak hasil perhitungan ùë¶(3.0)
print(f"Hasil Runge-Kutta orde 4: ùë¶(3.0) = {y}")

# Membuat Grafik Numerik
x_values = np.arange(x0, x_target + h, h)
y_values = [runge_kutta_4(x0, y0, h, x) for x in x_values]

# Menampilkan Hasil Grafik Numerik
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label="Runge-Kutta Orde 4", color="yellow")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hasil Grafik Orde 4")
plt.legend()
plt.grid(True)
plt.show()
```

2. Orde 4

```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan diferensial ùëëùë¶/ùëëùë• = 2
def f(x, y):
    return y**2 + 3

def runge_kutta_4(x0, y0, h, x_target):
    x = x0
    y = y0
    while x < x_target:
        k1 = h * f(x, y)
        k2 = h * f(x + h/2, y + k1/2)
        k3 = h * f(x + h/2, y + k2/2)
        k4 = h * f(x + h, y + k3)
        y = y + (k1 + 2*k2 + 2*k3 + k4) / 6
        x = x + h
    return y

# Awal
x0 = 0.0            # Nilai awal x
y0 = 0.0            # Nilai awal y
x_target = 0.2      # Target x
h = 0.1             # Ukuran langkah

# Menghitung ùë¶(0.2) dengan Runge-Kutta orde 4
y_rk4 = runge_kutta_4(x0, y0, h, x_target)

# Mencetak hasil perhitungan ùë¶(0.2)
print(f"Hasil Runge-Kutta orde 4: ùë¶(0.2) = {y_rk4}")

# Membuat Grafik Numerik
x_values = np.arange(x0, x_target + h, h)
y_rk4_values = []

for x in x_values:
    y_rk4_values.append(runge_kutta_4(x0, y0, h, x))

# Menampilkan Hasil Grafik Numerik
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_rk4_values, label="Runge-Kutta Orde 4", color="purple")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hasil Grafik Orde 4")
plt.legend()
plt.grid(True)
plt.show()
```
